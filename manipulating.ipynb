## Introduction
Images are stored as an array of pixels, each with its RGB value. This script goes over retrieving these values, and setting them. We will be working with an image of an A-10.

```python
!wget https://github.com/pandaero/OpenCV-101/raw/main/a-10.jpg
```

The script needs to load the necessary libraries, and "read" the image.

```python
# Import necessary libraries
%matplotlib inline
from matplotlib import pyplot as plt
import cv2

# Load the image
image = cv2.imread("a-10.jpg")

# Save and show the dimensions of the image
(h, w) = image.shape[:2]
print("Width: {}".format(w))
print("Height: {}".format(h))

# Show the image
plt.imshow(image)
```

Now we access a pixel to get its RGB value.

```python
# Access a pixel and save its value
(b, g, r) = image[500, 500]

# Print result
print("Pixel 500, 500: R {}, G {}, B {}".format(r, g, b))
```

We can assign this pixel a value.

```python
# Assign the pixel the colour red
image[500, 500] = (0, 0, 255)

# Get the pixel's value to confirm
(b, g, r) = image[500, 500]

# Show result
print("Pixel 500, 500: R: {}, G: {}, B: {}".format(r, g, b))
```

Now we can also get a *range* of pixels.

```python
# Get the centre of the image to split into quadrants
(cx, cy) = (w // 2, h // 2)

# Fetch the top-left quadrant
tlq = image[0:cy, 0:cx]

# Show the top-left quadrant
plt.imshow(tlq)
```

Likewise, we can assign a *range* of pixels.

```python
# Assigning the top-left quadrant the colour red
image[0:cy, 0:cx] = (255, 0, 0)

# Show the resulting image
plt.imshow(image)
```
